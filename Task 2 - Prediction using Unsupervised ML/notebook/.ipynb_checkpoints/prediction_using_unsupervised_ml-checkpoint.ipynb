### 4. Data Preparation

```python
X = df.drop('class',axis=1)
y = df['class'].map({'Iris-setosa':0 , 'Iris-versicolor':1, 'Iris-virginica':2 })
y= np.array(y)
```

```python
X.shape,y.shape
```

### 5. Model Building

>Here we have the target variable Class. We need to remove the target variable so that this dataset can be used to work
in an unsupervised learning environment. The iloc function is used to get the features we require.
We also use '.values' function to get an array of the dataset.
**(Note that we transformed the dataset to an array so that we can plot the graphs of the clusters)**.

```python
x = df.iloc[:, [0, 1, 2, 3]].values
```

#### Finding optimum number of clusters for K Means
#### Elbow Method using within-cluster-sum-of-squares(wcss)

```python
#  Kmeans is a kind of Unsupervised type of Clustering . It basically takes input from Dataset and predicts the clusters accordingly

from sklearn.cluster import KMeans

#Elbow Method
# Implicit objective function in k-Means measures sum of distances of observations from their cluster centroids,
called Within-Cluster-Sum-of-Squares (WCSS) 

wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', 
                    max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

### 6. Ploting Elbow Method

```python
plt.figure(figsize=(15,6))    
plt.plot(range(1, 11), wcss)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') # Within cluster sum of squares
plt.show()
```

```python
plt.figure(figsize=(15,6))
plt.scatter(range(1, 11), wcss,marker="*",c="green")       
plt.plot(range(1, 11), wcss,c="red",linestyle='dotted')
plt.xlabel("Number of clusters")           
plt.ylabel("WCSS")
plt.xticks(range(0,12,1))       #for value of k from 0 to 11 in graph
plt.title("Elbow curve") 

#Hence at k=3 , there is a steep decrease in Elbow curve, so this is best value of K
```

> we can clearly see why it is called "the elbow mehtod" from the above graph, the optimum clusters is where the elbow occurs. This is when the within cluster sum of squares (WCSS) doesn't decrease significantly with every iteration.
**In the above graph, it is there is no bend after 3rd cluster. Hence take number of clusters is 3**

### 7. Applying KMeans to the dataset

```python
k_means = KMeans(n_clusters = 3,init = 'k-means++',max_iter = 300, n_init=10, random_state=0,precompute_distances='auto', tol=0.0001, verbose=0)
y_km = k_means.fit_predict(x)
```

### 8. Predicted values

```python
y_km
```

### 10. Comparing Actual and Predicted values

```python
new_labels = k_means.labels_
# Plot the identified clusters and compare with the answers
fig, axes = plt.subplots(1, 2, figsize=(14,6))
axes[0].scatter(x[:, 0], x[:, 1], c=y, cmap='gist_rainbow',
edgecolor='k', s=50)
axes[1].scatter(x[:, 0], x[:, 1], c=new_labels, cmap='jet',
edgecolor='k', s=50)
axes[0].set_xlabel('Sepal length', fontsize=15)
axes[0].set_ylabel('Sepal width', fontsize=15)
axes[1].set_xlabel('Sepal length', fontsize=15)
axes[1].set_ylabel('Sepal width', fontsize=15)
axes[0].tick_params(direction='in', length=10, width=5, colors='k', labelsize=10)
axes[1].tick_params(direction='in', length=10, width=5, colors='k', labelsize=10)
axes[0].set_title('Actual', fontsize=18)
axes[1].set_title('Predicted', fontsize=18)
```

### 11. Finding out Cluster centers

```python
KModel = k_means.fit(x)
KModel
```

```python
KModel.labels_
```

```python
KModel.cluster_centers_
```

### 12. Plotting 2D clusters

```python
plt.figure(figsize=(10,6))
plt.scatter(x[y_km==0,0], x[y_km==0,1],s = 60,c = 'red', label = 'Iris-setosa' ,marker="*")
plt.scatter(x[y_km==1,0], x[y_km==1,1],s = 60,c = 'blue', label = 'Iris-versicolour' ,marker="*")
plt.scatter(x[y_km==2,0], x[y_km==2,1],s = 60,c = 'green', label = 'Iris-virginica' ,marker="*")
plt.scatter(k_means.cluster_centers_[:,0], k_means.cluster_centers_[:,1], s = 80, c = 'black', label = 'Centroid')
plt.title("Cluster of Species")
plt.legend()
plt.show()
```

### 13. Plotting 3D clusters

```python
fig = plt.figure(figsize = (15,15))
ax = fig.add_subplot(111, projection='3d')
plt.scatter(x[y_km == 0, 0], x[y_km == 0, 1], s = 100, c = 'purple', label = 'Iris-setosa')
plt.scatter(x[y_km == 1, 0], x[y_km == 1, 1], s = 100, c = 'orange', label = 'Iris-versicolour')
plt.scatter(x[y_km == 2, 0], x[y_km == 2, 1], s = 100, c = 'green', label = 'Iris-virginica')

#Plotting the centroids of the clusters
plt.scatter(k_means.cluster_centers_[:, 0], k_means.cluster_centers_[:,1], s = 300, c = 'black', label = 'Centroids')
plt.show()
```
